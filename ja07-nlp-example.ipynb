```python
from janome.tokenizer import Tokenizer
import pandas as pd

df = pd.read_csv('data/courses.csv')
df
```

```python
desc = df.description[0]  # df['description'][0] としてもよい
desc
```

```python
desc.replace('「', ' ').replace('」', ' ')
```

```python
table = str.maketrans({c: ' ' for c in '().,-「」（）、。●・\n'})
desc = desc.translate(table)
desc
```

```python
df.description = df.description.apply(lambda s: s.translate(table))
df
```

```python
tokenizer = Tokenizer()
for token in tokenizer.tokenize(desc):
    if token.surface !=  ' ':
        print(token)
```

```python
tokenizer = Tokenizer(wakati=True)
print(tokenizer.tokenize(desc))
```

```python
def tokenize(text):
    return [token for token in tokenizer.tokenize(text) if token != ' ']

tokenize(desc)
```

```python
stopwords = ['て', 'に', 'を', 'は', 'の', 'が', 'で', 'と', 'こと', 'できる', 'ある']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(tokenizer=tokenize, stop_words=stopwords, min_df=3)
X = cv.fit_transform(df.description)
```

```python
type(X), X.shape
```

```python
cv.get_feature_names()
```

```python
y = df.label
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,
                                                    random_state=789)
X_train.shape, X_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

model = LogisticRegression(C=0.1)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
model.classes_
```

```python
coefs = sorted(zip(model.coef_[0], cv.get_feature_names()))
coefs[:10]
```

```python
coefs[-10:]
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(n_estimators=100, min_samples_leaf=3)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
sorted(zip(model.feature_importances_, cv.get_feature_names()))[-10:]
```
